```python
from graphic_designer_tool.core.image_generator import ImageGenerator
from graphic_designer_tool.core.config import load_config
from IPython.display import Image, display
```

```python
config_path = "../configs/params.yml"
visual_description = "The Eiffel Tower on a rainy and foggy day",
```

```python
params = load_config(config_path=config_path)
```

```python
image_generator = ImageGenerator(
    n=params["n_image"], 
    dim=params["image_dim"],
    model=params["image_model"],
    quality=params["image_quality"],
)
```

```python
images = image_generator.get_images(visual_description=visual_description)
```
